```r
da=data.frame("name"=c("smith","vishnu","raj","hoi"),
              "bb"=c(22,44,55,66),
              "blood"=c("a","b","c","b-"))
```

```r
da
```

```r
library(reshape2)
```

melt

```r
#we need to pass the identity variables like name (discrete values) and measuremnet values
df=melt(da,id.vars=c("name"),measure.vars=c("bb","blood"))
```

```r
df
#changed (one single value)
```

cast

```r
df2=dcast(df,variable~name,value.var="value")
```

```r
df2
```

```r
# melt + cast
recast(df,(variable~name),id.var=c("name"))
```
